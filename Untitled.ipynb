```python
import time
from datetime import timedelta
from datetime import datetime

import pandas as pd
df=pd.read_csv("CB68.csv")

df1=pd.read_csv("lstm.csv")
date=df['Date'].astype('datetime64[ns]')
lstm_date=df1['Date'].astype('datetime64[ns]')

device=df['Light1']
lstm=df1['status']
total_prewards=0
total_nrewards=0

new=[]
#print(len(date))
for i in range (len(lstm_date)):
    prewards=0
    nrewards=0
    count=0
    count1=0
    for j in range (len(date)):
        #print("data:",date[j])
        if(date[j]==lstm_date[i]):
            next_minute= lstm_date[i]+ timedelta(minutes=15)
            #print("next:",next_minute)
            
            k=j
            for k in range (j,len(date)):
                #print(k)
                
                if(date[k]>=next_minute): break
                if(lstm[i]==device[k]):
                    count1=0
                    count=count+1
                    if(count==1):
                       
                        prewards=prewards+1
                else:
                    count=0
                    count1=count1+1
                    if(count1==1):
                        nrewards=nrewards+1
                        print("overiding happened at:",date[k],device[k])
                        new.append(date[k])
            
    
    print(i+1,"lstm schedule_rewards_______________:")
    print("total prewards:",prewards)
    print("total nrewards:",nrewards)
    total_prewards=prewards+total_prewards
    total_nrewards=nrewards+total_nrewards
print("________totalrewards______________")
print("total prewards:",total_prewards)
print("total nrewards:",total_nrewards) 
#print(new)

def checkList( list):
    first = list[0]
    n=len(list)
    if(list.count(first) == len(list)):
        print("regularly overding at :",list[n-1])
    else:
        print("irregularly overiding")

checkList(new)


        
    
 

max_value = None

for num in new:
    if (max_value is None or num > max_value):
        max_value = num

print('Maximum value:', max_value)       


    
```
